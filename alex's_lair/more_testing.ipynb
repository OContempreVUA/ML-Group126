```python
df = pd.read_csv("some_dataset.csv", index_col=0)
df.columns = ["x-axis", "y-axis", "color"]
df.head()
# print(df.values)
data = df.values
df.head()
```

```python
red = df.loc[df["color"] == "red"]
green = df.loc[df["color"] == "green"]

red_data = red.values
green_data = green.values
```

```python
import matplotlib.pyplot as plt

# Sample data
# Plotting the data
plt.scatter(red_data[:,0], red_data[:,1], marker='o', linestyle='-', color="red")  # Plot data points with lines connecting them
plt.scatter(green_data[:,0], green_data[:,1], marker="o", color="green")
plt.title('Sample 2D Data')  # Add title to the plot
plt.xlabel('X-axis')  # Add label to the x-axis
plt.ylabel('Y-axis')  # Add label to the y-axis
plt.grid(True)  # Optional: add gridlines to the plot
plt.show()  # Show the plot
```

```python
import matplotlib.pyplot as plt
n_points: int = 50
new_dataset = np.random.randint(low = 0, high = 100, size = [n_points,2])

colors: list = []
for _ in range(1,n_points+1):
    random_num = np.random.random()
    if random_num < 0.5:
        colors.append("red")
    else:
        colors.append("green")
        
new_dataset = np.column_stack((new_dataset, colors))
new_df: pd.DataFrame = pd.DataFrame(data=new_dataset, columns=["x-axis", "y-axis", "color"])
new_df.head()

new_red_df = new_df.loc[new_df["color"] == "red"]
new_red_data = new_red_df.values
new_green_df = new_df.loc[new_df["color"] != "red"]
new_green_data = new_green_df.values

plt.scatter(new_red_data[:,0], new_red_data[:,1], marker='o', linestyle='-', color="red")  # Plot data points with lines connecting them
plt.scatter(new_green_data[:,0], new_green_data[:,1], marker="o", color="green")
plt.title('Sample 2D Data')  # Add title to the plot
plt.xlabel('X-axis')  # Add label to the x-axis
plt.ylabel('Y-axis')  # Add label to the y-axis
plt.grid(visible=True)  # Optional: add gridlines to the plot
plt.show()  # Show the plot
```
